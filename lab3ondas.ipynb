```julia
using Plots
function flower(k, n, p=30)
    r(θ) = (k*sin(n*θ)+p)/p
    x(θ) = r(θ)*cos(θ)
    y(θ) = r(θ)*sin(θ)
    θ = 0:0.01:2π
    return x.(θ), y.(θ)
end

function espiras(armonico)
    t1 = collect(-4:0.5:4)
    t = append!(t1, collect(4:-0.5:-4))
    t2 = collect(4:-0.5:-4)
    t_ = append!(t2, collect(-4:0.5:4))
    g = @gif for (i, j) in zip(t, t_)
        x1, y1 = flower(i, 2*armonico+1)
        x2, y2 = flower(j, 2*armonico+1)
        plot(x1, y1, aspect_ratio=1, ylims=(-1.2,1.2), lw = 2, legend=:topright, label = false)
        plot!(x2, y2, linestyle=:dash, lw = 2, legen=:topright, label = false)
        annotate!(0, 0, text("Armónico #$armonico"), fontsize = 3)
    end fps = 30
    display(g)
end
```

```julia
espiras(1)
```

```julia
espiras(2)
```

```julia
espiras(3)
```
